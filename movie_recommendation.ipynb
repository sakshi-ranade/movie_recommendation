```python
class Movie(): #creating a movie class
    def __init__(self, title,release_date, genres, vote_average):
        self.title = title
        self.release_date = release_date
        self.genres = genres
        self.vote_average = vote_average
        #self.popularity = popularity
        #self.spoken_languages = spoken_languages
        #self.production_companies = production_companies
    def display(self):
        print(f"{self.title}, released in {self.release_date}, Genre: {self.genres}, Rating: {self.vote_average}")
```

```python
movie_data = []
for index, row in df.iterrows():
    movie = Movie(row['title'], row['release_date'], row['genres'], row['vote_average'])
    movie_data.append(movie)
```

```python
class Database():
    def __init__(self,df):
        self.movie_data = self.get_movie_data(df)

    def get_movie_data(self,df):
        movie_data = []
        for index, row in df.iterrows():
            movie = Movie(row['title'], row['release_date'], row['genres'], row['vote_average'])
            movie_data.append(movie)
        return movie_data

    def recommend_by_rating(self, min_rating):
        print(f"\nMovie with ratings above {min_rating}: ")
        for movie in self.movie_data:
            if movie.vote_average >= min_rating:
                movie.display()

    def recommend_by_genre(self, *genre):
        print(f"\nMovies with all genres: {', '.join(genre)}")
        for movie in self.movie_data:
        # Check if the movie contains all the genres specified by the user
            if all(g in movie.genres for g in genre):
                movie.display()
```

```python
movies = Database(df)
```

```python
movies.recommend_by_rating(9.1)
```

```python
movies.recommend_by_genre('Drama', 'Action')
```
