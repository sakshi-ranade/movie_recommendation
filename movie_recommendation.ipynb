```python
import pandas as pd
```

```python
df = pd.read_csv('imdb_movies-1.csv')
```

```python
df.head()
```

```python
class Movie(): #creating a movie class
    def __init__(self, title,release_date, genres, vote_average):
        self.title = title
        self.release_date = release_date
        self.genres = genres
        self.vote_average = vote_average
        #self.popularity = popularity
        #self.spoken_languages = spoken_languages
        #self.production_companies = production_companies
    def display(self):
        print(f"{self.title}, released in {self.release_date}, Genre: {self.genres}, Rating: {self.vote_average}")
```

```python
df.columns
```

```python
movie_data = []
for index, row in df.iterrows():
    movie = Movie(row['title'], row['release_date'], row['genres'], row['vote_average'])
    movie_data.append(movie)
```

```python
movie_data[0].display()
```

```python
min_rating = 7.8
```

```python
def recommend_by_rating(movie_data, min_rating):
    print(f"\nMovie with ratings above {min_rating}: ")
    for movie in movie_data:
        if movie.vote_average >= min_rating:
            movie.display()

#calling the recommend by rating function                
recommend_by_rating(movie_data, min_rating)
```
